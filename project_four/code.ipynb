```python
def int2hex(element):
    temp = hex(element)
    if temp[-1] == 'L':
        temp = temp[:-1]
    if temp[0:2] == '0x':
        return temp[2:]
    else:
        return temp
```

```python
def int2bin(element):
    return bin(element)[2:]
```

```python
def bin2int(element):
    return int(element,2)
```

```python
def Xor(str1,str2,b):
    y = int(str1,2) ^^ int(str2,2)
    return STRB(y,b)
```

```python
def STRB(input_int,b):
    int_to_bin = bin(input_int)[2:]
    num_zeros = b - len(int_to_bin)
    int_to_bin = ('0'* num_zeros) + int_to_bin
    return int_to_bin
```

```python
def decode_point(point):
    
    zero_byte = hex2byte('0' + int2hex(0))
    is_infinity = True
    for i in range(0,len(point)):
        if point[i] != zero_byte:
            is_infinity = False
            break
    if is_infinity:
        return EC(0,0)
    else:
        last_byte = point[-1]
        if ((last_byte)!= hex2byte(int2hex(15))) and ((last_byte)!= hex2byte(int2hex(255))):
            raise Exception('invalid last encoding , byte = {}'.format(last_byte))
        else:
            x_coor_bytes = point[:-1]
            x_coor_int =  decode_field_element(x_coor_bytes)
            elliptic_points = EC.lift_x(x_coor_int,all=True)
            
            # select point based on parity
            point_zero = elliptic_points[0].xy()
            point_zero_y = int(point_zero[1])
            if last_byte == hex2byte(int2hex(15)):
                if is_odd(point_zero_y):
                    return elliptic_points[0]
                else:
                    return elliptic_points[1]
            else:
                if is_even(point_zero_y):
                    return elliptic_points[0]
                else:
                    return elliptic_points[1]
            
            
            return elliptic_point
```

```python
def encode_into_domain(x):
    
    x_hex = int2hex(x)
    x_hex = '0'* (8 - len(x_hex)) + x_hex # pad to get 4 bytes
    if len(x_hex)>8:
        x_hex = x_hex[-8:] # return last 8 bytes
    x_bytes = hex2byte(x_hex)
    return x_bytes[::-1] # reverse bytes
```

```python
def inverse_comp(P,K3,EC):
    
    first_two_bytes = P[0:2] # take first two bytes
    p_length = len(P[2:])/2
    
    # split p into half first for alpha and second for beta
    alpha = Integer(hex2int(P[2:p_length+2]))
    beta = Integer(hex2int(P[p_length+2:]))
    
    if first_two_bytes == '00':
        return EC(int2hex(0))
    
    P = EC.lift_x(K3([alpha,beta])) # compute point
    
    _,Py = P.xy()
    gamma,_ = Py.polynomial().coefficients()
    
    # take point based on parity
    if ((int(gamma)%2==0) and first_two_bytes == '03') or (not (int(gamma)%2==1) and first_two_bytes == '02'):
        P  = EC(K3([alpha,beta]),-Py)
    
    return P
```

```python
def permuteUnique(nums):
    
        # a function to find all permutations of a string
    
        def backtrack(tmp, size):
            if len(tmp) == size:
                ans.append(tmp[:])
            else:
                for i in range(size):
                    if visited[i] or (i > 0 and nums[i-1] == nums[i] and not visited[i-1]):
                        continue
                    visited[i] = True
                    tmp.append(nums[i])
                    backtrack(tmp, size)
                    tmp.pop()
                    visited[i] = False
        ans = []
        visited = [False] * len(nums)
        nums.sort()
        backtrack([], len(nums))
        return ans
```

```python
def send_message(connection,msg):
    
    # a function in which we send a message to the server and return
    # the response as a string
    
    connection.send(msg + "\n")
    response=""
    while True: #data might come in several packets, need to wait for all of it
        data = connection.recv(1024)
        if len(data) == 0:
            break
        if data[-1] == '\n': 
            response += data[:-1]  
            break
        response += data
    return response
```

```python
def query_for_AB(server_name, port, message):
    
    # send a query to the server and then compute A and b from the response received
    
    server = (server_name, int(port)) 
    s = socket.create_connection(server)
    response = send_message(s,message)
    
        
    response_parsed = response.split(' ')
    A_base,b_base = response_parsed[0],response_parsed[1]
    # change a and b to hex from base64
    A_hex,b_hex = binascii.hexlify(base64.decodestring(A_base)),binascii.hexlify(base64.decodestring(b_base))
    b_bin = hex2bin(b_hex) # change b to binary
    A_bin = hex2bin(A_hex) # change A to binary
    A_mtrx = []
    # change A from string format to matrix format
    for i in range(0,16384,128):    
        A_mtrx.append(list(map(lambda x : int(x) ,list(A_bin[i:i+128]))))
    
    return A_mtrx,b_bin,s
    
    
```

```python
def compute_s_zero(A,ms,b):
    
    A_inverse = matrix(GF(2),128,A).inverse() # compute inverse of matrix A
    a_s_zero = list(map(lambda x : int(x) ,list(Xor(ms,b,128)))) # xor each vector in ms and b
    a_s_zero = matrix(GF(2),1,a_s_zero) # change the list into GF vector
    s_zero  = A_inverse*a_s_zero.transpose() # compute the dot product
    return s_zero.numpy() # return s_zero
```

```python
def decrypt(server_name, port, message , possible_ms):
    
    s_zero_key = "0" * 128
    
   
    AES_zero = AES.new(bin2byte(s_zero_key))
    
    
    # keep iterating until you find a full rank A
    A,b,con = query_for_AB(server_name, port, message )    
    ms = possible_ms[0] 
    while True:

        try:
             # given response calculate As0 and then we multiply with a inverse to get s zero
            s_zero = compute_s_zero(A,ms,b)
            break
        except Exception as e:
            con.close()
            A,b,con = query_for_AB(server_name, port, message )

            
    # try all possible messages of having one 1 , find message that is readable
    for i in range(0,len(possible_ms)): 
        
        ms = possible_ms[i]
        s_zero = compute_s_zero(A,ms,b)
        
        # conversions to list of binary information
        conversion_np = s_zero.reshape(-1).tolist()
        conversion_np = list(map(lambda x : str(x),conversion_np))
        s_zero_str = "".join(conversion_np)
        s_zero_byte = bin2byte(s_zero_str)

        # get possible password
        possible_pwd = AES_zero.decrypt(s_zero_byte)
        pwd_bin = byte2bin(possible_pwd)
        
        if pwd_bin[-1] != '1': # if password doesnt end with one since 0^(128 - /w/) can be at least one zero
        
            pwd_asci = util.bin_to_ascii(pwd_bin)
            if all(ord(c) < 128 for c in pwd_asci):
                print(pwd_asci)
                break
```
